```python
import pandas as pd

df1 = pd.read_csv('./review_위경도.csv', encoding='cp949')
df1
```

```python
df = df1[['name', 'review']]
df
```

```python
#!pip install KeyBERT
```

```python
def BERT(name):

    array_text = pd.DataFrame(df[df['name'] == name]['review']).to_numpy()

    bow = []
    from keybert import KeyBERT
    kw_extractor = KeyBERT('distilbert-base-nli-mean-tokens')
    for j in range(len(array_text)):
        keywords = kw_extractor.extract_keywords(array_text[j][0])
        bow.append(keywords)
    
    new_bow = []
    for i in range(0, len(bow)):
        for j in range(len(bow[i])):
            new_bow.append(bow[i][j])
            
    keyword = pd.DataFrame(new_bow, columns=['keyword', 'weight'])
    print(keyword.groupby('keyword').agg('sum').sort_values('weight', ascending=False).head(5))
```

```python
BERT('류몽민')
```

```python
BERT('세븐브릭스키친')
```
