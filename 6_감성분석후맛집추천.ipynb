```python
def score_evaluate(score):
    if score >= 3.5:
        return 1
    else:
        return 0
df['NP'] = df['score'].apply(score_evaluate) # 긍부정 여부 1,0으로 라벨링
```

정규표현식

```python
import re
```

```python
def text_clearing(text): 
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')
    # 지정한 정규식에 해당하지 않은 것은 길이가 0인 문자열로 변환
    result = hangul.sub('', text)
    return result
```

```python
df["ko_review"] = df["review"].apply(lambda x : text_clearing(x))
```

```python
df.drop("review", axis=1, inplace=True)
df
```

형태소분석

```python
from konlpy.tag import Okt
```

```python
stopwords = ['도','한','는','은','이','가','와','로','에','들','의','과','자','의','좀','잘','걍','으로','하다','님','다','를','을','자','랑','게','그','에서']
```

```python
okt = Okt()
```

```python
okt_review = []
for sentence in df['ko_review']:
    tokenized_sentence = okt.morphs(sentence, stem=True)
    stopwords_removed_sentence = [word for word in tokenized_sentence if not word in stopwords] # 불용어 제거
    okt_review.append(stopwords_removed_sentence)
```

```python
okt_review
```

토큰화 & 패딩

```python
from tensorflow.keras.preprocessing.text import Tokenizer 
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# 정수 인코딩 (텍스트를 숫자로 처리할 수 있도록 훈련, 테스트 데이터에 정수 인코딩) 
tokenizer = Tokenizer() 
tokenizer.fit_on_texts(okt_review)

# 단어 집합 생성 -> 높은 숫자일수록 등장 빈도수가 낮음. 
# 등장 빈도수 낮은애들 제거 
word_min = 2 
total_count = len(tokenizer.word_index) 
rare_count = 0 # 등장이 2회 미만

# 빈도수 VALUE 
for key,value in tokenizer.word_counts.items(): 
    if value < word_min: 
        rare_count = rare_count + 1
```

```python
vocab_size = total_count - rare_count + 2 # OOV와 padding 때문에 +2
```

```python
vocab_size
```

```python
tokenizer = Tokenizer(vocab_size, oov_token='OOV') # OOV 토큰 
tokenizer.fit_on_texts(okt_review)
```

```python
word_dic = tokenizer.word_index
word_dic
```

```python
sequences = tokenizer.texts_to_sequences(okt_review)
print(sequences)

padded = pad_sequences(sequences)
print(padded)
```

학습과 테스트 세트 분리

```python
x = padded
y = df['NP']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
print(x_train.shape)
print(x_test.shape)
```

LSTM

```python
from tensorflow.keras.layers import Embedding,Dense,LSTM
from tensorflow.keras.models import Sequential 
from tensorflow.keras.models import load_model 
from tensorflow.keras.callbacks import EarlyStopping,ModelCheckpoint
```

```python
model = Sequential() 
model.add(Embedding(vocab_size,64)) 
model.add(LSTM(32)) 
model.add(Dense(1,activation='sigmoid'))
```

```python
es = EarlyStopping(monitor='val_loss', mode = 'min', verbose = 1, patience=4) 
mc = ModelCheckpoint('best_model.h5', monitor='val_loss', mode='max', verbose=1, save_best_only=True)
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc']) 
history = model.fit(x_train, y_train, epochs=100, callbacks=[es,mc], batch_size=200, validation_split=0.2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def show_history(history):
  plt.figure(figsize=(6, 6))
  plt.yticks(np.arange(0, 1, 0.05))
  plt.plot(history.history['acc'], label='train')
  plt.legend()

show_history(history)
```

```python
loaded_model = load_model('best_model.h5') 
print("정확도 : %.4f" % (loaded_model.evaluate(x_test, y_test)[1]))
```

### 강남구 리뷰 분석

```python
df1 = pd.read_csv("./서울_강남구.csv")

df1
```

```python
df1.info()
```

```python
df1 = df1.dropna()
df1.info()
```

```python
df1['review']=df1['review'].str.replace(pat=r'[^ㄱ-ㅣ가-힣]+', repl= r'', regex=True)
df1['review']
```

```python
filter = df1['review'] != ""
review_data = df1[filter]
review_data
```

```python
def sentiment_predict(new_sentence): 
    new_sentence = re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣 ]','', str(new_sentence))
    new_sentence = okt.morphs(new_sentence, stem=True) # 토큰화
    new_sentence = [word for word in new_sentence if not word in stopwords] # 불용어 제거
    encoded = tokenizer.texts_to_sequences([new_sentence]) # 정수 인코딩
    pad_new = pad_sequences(encoded) # 패딩
    score = float(loaded_model.predict(pad_new)) # 예측
    if(score > 0.5):
        ans = 1
    else:
        ans = 0
    return ans
```

```python
test_sentences = review_data['review'].to_list()
score = []
cnt = 0

for sentences in test_sentences:
    try:
        np = sentiment_predict(sentences)
        score.append(np)
        cnt += 1
    except:
        print('에러문장:',sentences)
        score.append('error')
        cnt += 1
    if cnt % 100 == 0:
        print(f'현재{cnt}개 진행완료')
```
